```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
from scipy.stats import lognorm
from scipy.stats import weibull_min
from scipy.stats import uniform
from scipy.stats import norm
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

plt.style.use(['ager'])
#plt.style.use(['ieee'])

# uniformation distribution
loc_u = 2.5e-5
scale_u = 2.5e-5
x_u = np.linspace(uniform.ppf(0.01, loc=loc_u, scale=scale_u),
                  uniform.ppf(0.99, loc=loc_u, scale=scale_u), 100)
u1 = uniform.rvs(size=1000, scale=scale_u, loc=loc_u)

# normal distribution
loc_n = (2.5e-5 + 5.0e-5) / 2
scale_n = 3.5e-5 / (norm.ppf(0.99) - norm.ppf(0.01))

x_n = np.linspace(norm.ppf(0.01, scale=scale_n, loc=loc_n),
                  norm.ppf(0.99, scale=scale_n, loc=loc_n), 100)
n1 = norm.rvs(size=1000, scale=scale_n, loc=loc_n)

# weibull distribution
c = 1.7
loc_w = 2.5e-5
scale_w = 2.5e-5 / (weibull_min.ppf(0.99, c) - weibull_min.ppf(0.01, c))
x_w = np.linspace(weibull_min.ppf(0.01, c, scale=scale_w, loc=loc_w),
                  weibull_min.ppf(0.99, c, scale=scale_w, loc=loc_w), 100)
w1 = weibull_min.rvs(c=c, size=1000, scale=scale_w, loc=loc_w)

# log normal distribution
s = 0.5
loc_l = 2.5e-5
scale_l = 2.5e-5 / 4
x_l = np.linspace(lognorm.ppf(0.01, s, scale=scale_l, loc=loc_l),
                  lognorm.ppf(0.99, s, scale=scale_l, loc=loc_l), 100)
g1 = lognorm.rvs(s=s, size=1000, scale=scale_l, loc=loc_l)

# plot
fig1, axs = plt.subplots(2, 2)

axs[0, 0].hist(u1, density=True, histtype='stepfilled',
               color='b', alpha=0.2, label='random diameters')
axs[0, 0].plot(x_u, uniform.pdf(x_u, loc=loc_u, scale=scale_u),
               'r-', lw=5, alpha=0.6, label='uniform pdf')

axs[0, 1].hist(n1, density=True, histtype='stepfilled',
               color='b', alpha=0.2, label='random diameters')
axs[0, 1].plot(x_n, norm.pdf(x_n, loc=loc_n, scale=scale_n),
               'r-', lw=5, alpha=0.6, label='normal pdf')


axs[1, 0].hist(w1, density=True, histtype='stepfilled',
               color='b', alpha=0.2, label='random diameters')
axs[1, 0].plot(x_w, weibull_min.pdf(x_w, c, loc=loc_w, scale=scale_w),
               'r-', lw=5, alpha=0.6, label='weibull pdf')
axs[1, 1].hist(g1, density=True, histtype='stepfilled',
               color='b', alpha=0.2, label='random diameters')
axs[1, 1].plot(x_l, lognorm.pdf(x_l, s, loc=loc_l, scale=scale_l),
               'r-', lw=5, alpha=0.6, label='lognorm pdf')

axs[0, 0].set_xlabel('uniform distribution pore diameters (m)', fontsize=24)
axs[0, 0].set_ylabel('normalized frequency', fontsize=24)
#axs[0, 0].set_title('', fontsize=24)
axs[0, 0].tick_params(labelsize=24)
axs[0, 0].legend(prop={'size': 20, }, loc='upper right')
axs[0, 0].xaxis.get_major_formatter().set_powerlimits((-6, -5))
axs[0, 0].yaxis.get_major_formatter().set_powerlimits((3, 4))
axs[0, 0].tick_params(axis='x', which='major', reset=False, length=5)
axs[0, 0].tick_params(axis='both', which='both', reset=False, length=5)


axs[0, 1].set_xlabel('normal distribution pore diameters (m)', fontsize=24)
axs[0, 1].set_ylabel('normalized frequency', fontsize=24)
#axs[0, 1].set_title('normal distribution', fontsize=24)
axs[0, 1].tick_params(labelsize=24)
axs[0, 1].legend(prop={'size': 20, }, loc='upper right')
axs[0, 1].xaxis.get_major_formatter().set_powerlimits((-4, -5))
axs[0, 1].yaxis.get_major_formatter().set_powerlimits((3, 4))

axs[1, 0].set_xlabel('weibull distribution pore diameters (m)', fontsize=24)
axs[1, 0].set_ylabel('normalized frequency', fontsize=24)
#axs[1, 0].set_title('weibull distribution', fontsize=24)
axs[1, 0].tick_params(labelsize=24)
axs[1, 0].legend(prop={'size': 20, }, loc='upper right')
axs[1, 0].xaxis.get_major_formatter().set_powerlimits((-4, -5))
axs[1, 0].yaxis.get_major_formatter().set_powerlimits((3, 4))

axs[1, 1].set_xlabel('lognorm distribution pore diameters (m)', fontsize=24)
axs[1, 1].set_ylabel('normalized frequency', fontsize=24)
#axs[1, 1].set_title('lognorm distribution', fontsize=24)
axs[1, 1].tick_params(labelsize=24)
axs[1, 1].legend(prop={'size': 20, }, loc='upper right')
axs[1, 1].xaxis.get_major_formatter().set_powerlimits((-4, -5))
axs[1, 1].yaxis.get_major_formatter().set_powerlimits((3, 4))

#plt.savefig('F:\\xu\\python\\jupyter\\pnm\\random_pore_size_distribution\\random_pore_size_distribution.png')
plt.show()
```
